```python
# # 파일 종류 : 
# # - 바이너리(숫자로 되어있다.) -> 인코딩, 디코딩 둘다 일어나지 않음
# # --> 이미지, 동영상, 오디오, 실행파일 등
# # --> 컴퓨터가 알아볼 수 있는 문자로만 되어 있음. 
# # - 텍스트 파일(문자가 저장된 파일)로 구분 , 문자로 처리 : 인코딩, 디코딩
# # --> 문자로 볼 수 있다.(인코딩, 디코딩로 눈으로 확인 가능함)
# # 메모장에서 텍스트 파일을 열면 화면에 문자가 보인다.(디코딩)
# # 파일 생성 : 
# # 모드 결정(텍스트,바이너리) rwa 둘 다 가능
# # 읽기(r),덮어쓰기(w),추가쓰기(a) 모드 결정 
# # 덮어쓰기(1.이미 존재시 , 2.신규파일일 경우)
# # 파일 생성 : 텍스트 모드, w, 해당파일이 없는 경우(생성)
# # 텍스트 파일을 생성 
# # -> 파일명 : 새파일 , 확장자 : txt, 위치 : chap04 동일 위치(상대,절대)
# # 1. open
# # 2. write
# # 3. close
# 상대 경로
f = open(
    '새파일.txt', # file 매개변수 값을 지정, 경로와 파일명 확장자(경로가 없으면 현재 위치)
    'w'          # 텍스트의 쓰기 모드 지정 
)

f.close() # 파일 닫기(메소드는 점으로 접근한다)

# # 절대 경로 방식
# # f = open(
# #     'C:\big21\python_dev\chap04\새파일.txt', # file 매개변수 값을 지정, 경로와 파일명 확장자(경로가 없으면 현재 위치)
# #     'w'          # 텍스트의 쓰기 모드 지정 
# # )

# # with open('example.txt', 'w') as file:
# #     file.write("Hello, world!")
```

```python
# 파일을 쓰기 모드로 열어 출력값 적기
# f -> w(object) -> 새파일.txt 접근
# write 할때 버퍼에 저장이 됨
# 10줄을 파일에 쓰기
# 윈도우는 cp949
# 파이썬 자체는 utf-8(유니코드) 사용함
# 윈도우는 기본으로 cp949로 하기 때문에 오류가 발생함.
# 파이썬에서 할때는 utf-8로 진행해야 한다.
# 파일 읽고,쓰기는 다른 언어에 비해 편하다
# close 안하면 다른 사람이 사용할수가 없다.
f = open('./새파일.txt', 'w', encoding="utf-8") # 파일 열기
for i in range(1, 11): # 1부터 10까지 반복
    data = f'{i}번째 줄입니다.\n' # 파일에 쓰여질 데이터 저장
    f.write(data) # 파일에 문자열 쓰기
f.close() # 파일 닫기
```

```python
# 파일 읽기 : 파일이 존재,없으면 에러
#  운영체제의 인코딩/디코딩 방식을 불러서 디폴트로 그 방식으로 처리한다
f = open('./새파일.txt', 'r', encoding="utf-8") # 읽을수 있는 오브젝트 생성한 것이다.
# read 1줄로 전체읽음
# readline 1줄씩
# readlines 1줄에 읽되[리스트로 읽음]
line = f.readline() # 파일에서 한 줄 읽기
print(line) # 읽은 줄 출력
f.close() # 파일 닫기

```

```python
# 여러줄 읽어서 출력
f = open('./새파일.txt', 'r', encoding="utf-8") # 읽을수 있는 오브젝트 생성한 것이다.
for i in range(1, 11):
    line = f.readline()
    print(line.strip()) # .strip()으로 뒤에 붙은 엔터값(\n)을 깔끔하게 지워줍니다.
f.close() # 파일 닫기
```

```python
# readlines() : 파일 전체를 읽어서 리스트로 반환한다.
# ['1번째줄','2번째줄','3번째줄',~~~~~'10번째줄']
f = open('./새파일.txt', 'r', encoding="utf-8") # 읽을수 있는 오브젝트 생성한 것이다.
lines = f.readlines() # 파일 전체를 읽어서 리스트로 반환한다.
# print(lines[0])
for line in lines:
    print(line.strip()) # .strip()으로 뒤에 붙은 엔터값(\n)을 깔끔하게 지워줍니다.
f.close() # 파일 닫기


# f = open('./새파일.txt', 'r', encoding="utf-8") # 읽을수 있는 오브젝트 생성한 것이다.
# lines = f.readlines() # 파일 전체를 읽어서 리스트로 반환한다.
# print(lines) # 읽은 줄 출력
# f.close() # 파일 닫기
```

```python
# 텍스트 파일에 대한 예제였다.
# csv 파일 읽기 , 쓰기 하는 기능(함수)을 가지고 있는 모듈 : csv.py 모듈
# 설치는 되어 있는데, 외장모듈(인터프리터 안에 존재하지는 않음)
# 외장은 많은 툴이 설치가 된다.(인터프리터는 기본)
# 내장함수(인터프리터)
# 외장함수(설치시 import로 사용함)
# 모듈안에 1. 글로벌변수 2. 함수 3. 클래스 등이 존재한다.
# 함수란 : 특정 작업을 수행하는 코드 묶음
# 필요할 때마다 불러 쓸 수 있게 만든 것
# 어떻게 잘 쓸수 있는지 알수 있는것이 중요하다
# 기존에 어떤 기능이 알아야 그대로 사용하면 된다.
# 그게 바이브코드에 기본
import csv # csv 모듈을 가져와라, 외장이라서 이 명령이 필요하다

f = open('./data/data.csv', 'r', encoding="utf-8") # csv 파일을 읽기 모드로 열기
rdr = csv.reader(f) # csv 현태로 읽을 수 있는 구조 생성
for line in rdr: # rdr 객체에서 각 행을 반복하여 읽기
    print(line) # 각 행을 출력
f.close() # 파일 닫기
```

```python
# csv 파일에 쓰기
import csv
# data 디렉토리 output.csv
f = open('./data/output.csv', 'w', encoding="utf-8",newline='')
# csv write 할 수 있는 오브젝트를 생성
wr = csv.writer(f) # csv 형태로 쓸 수 있는 구조 생성                    
wr.writerow(['1','홍길동',False])
wr.writerow(['2','고길동',True])
f.close() # 파일 닫기

```

```python
# JSON 파일 입출력(요즘 데이터 주고 받을때 표준이라고 보면 된다)
# 자바스크립트에서 오브젝트를 표기하는 문법{KEY:VALUE}
# 파이썬의 딕셔너리 구조이기에 데이터 형태로 변환 사용 가능하다
# 이기종 데이터를 딕셔너리로 데이터를 처리할 수 있다
# JS(JSON) -> 파이썬(딕셔너리) -> JS(JSON) : JSON 모듈(외장)
# JSON : load(), dump() : JSON 파일 입출력 함수


```

```python

import json # json 모듈을 가져와라, 외장이라서 이 명령이 필요하다
데이터 = { #딕셔너리 생성
    '이름': '홍길동', 
    '나이': 30, 
    '취미': ['코딩','운동']
} # 딕셔너리 형태의 데이터 생성    

# json 형식으로 파일에 저장
with open('./data/프로필.json', 'w', encoding='utf-8') as f: # json 파일을 쓰기 모드로 열기
    json.dump(
        데이터,  #파일에 저장할 딕셔너리 지정
        f,      #파일   
        ensure_ascii=False, #한글이 깨지지 않고 그대로 저장하기 위해
        indent=2 #파일에 저장할 때 들여쓰기 설정, 가독성 향상(앞에 2칸 들여쓰기)       
    ) # 데이터를 json 형식으로 파일에 저장
```

```python
# json 형식의 파일 -> 파이썬 객체(딕셔너리) 읽기
with open('./data/프로필.json', 'r', encoding='utf-8') as f: # json 파일을 읽기 모드로 열기
    dict_data = json.load(f) # json 형식의 파일을 읽어서 파이썬 객체(딕셔너리)로 변환하여 저장     
print(dict_data.get('이름')) # 읽은 데이터 출력
print(dict_data.get('나이')) # 읽은 데이터 출력
print(dict_data.get('취미')) # 읽은 데이터 출력
```
