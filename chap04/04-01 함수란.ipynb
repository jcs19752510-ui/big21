```python
# 함수 선언과 실행
# 문제 : 두 정수를 입력받아 합을 구하는 함수(입력 -> 코드 -> 반환)
# 입력 : 두 정수
# 코드 : 두 정수를 더하기
# 반환 : 합
# def 함수이름(입력) : 
#   코드 함수 처리 기능
#   return
def sum_(int_a , int_b) :
    result = int_a + int_b
    return result

def sum__(int_a , int_b) :
    return int_a + int_b

sum_(10,20) , sum__(10,20)
```

```python
# 가변 인자(arguments)
# 2개 정수의 합을 구하는 함수 선언, 실행
def 합계2(x,y):
    total = x + y
    return total

# 3개 정수의 합을 구하는 함수 선언, 실행
def 합계3(x,y,z):
    total = x + y + z
    return total

합계2(1,2)
합계3(1,2,3)
```

```python
# 가변 인자 적용
def 합계(*numbers): # *numbers :  가변매체변수(들어올때는 튜플(인덱스)반복가능. 로 들어온다.)
    total = 0
    for number in numbers:
        total += number
    return total

print(합계(1,2))     #튜플
print(합계(1,2,3))   #튜플
print(합계(1,2,3,4)) #튜플
```

```python
# 가변 인자 : ** -> 개수가 정해지지 않은 키워드 인자(key) , 
# 함수 내부에서 딕셔너리 형태로 처리된다
# 여러개의 딕셔너리 값을 추출해 낼수 있다.
def 프로필(**정보): # **두개이면 딕셔너리로 처리하겠다. 의미
    for key, value in 정보.items():
        print(f'{key} : {value}')      
```

```python
# 함수의 문서화 : 독스트링(docstring), 함수에 대한 설명 적어 놓은것
# 함수를 잘 만들어 놓고 독스트링을 잘 만들어 놓으면
# AI 독스트링을 찾고하여 함수를 호출하게 된다.
# 구글 스타일로 기재하려고 한다.
# 용도와 문법이 따로 있다.
# 환율변환 함수
# 인공지능 관점에서 독스트링을 만드는 것이다.
# 즉, 독스트링은 인공지능을 위한 것이다.
# 이게 진정한 바이브 코딩이다.
# 아래 내용은 구글 스타일이다.(대부분 이 방식을 사용한다.)
# 독스트링은 무조건 첫줄에 와야한다
def 환률변환(원화:float, 환율:float) -> float: # :float 타입힌트
    """원화를 외화로 환산한다.

    Args : 
        원화 : 변환할 금액(원)
        환율 : 1외화당 원화 가격
    Returns :
        변환된 외화 금액
    """
    # 코드
    return 원화/환율
```

```python
# 독스트링 사용방법은 2가지 있다.
# print(환률변환.__doc__) # doc스트링을 출력한다.
help(환률변환) # doc스트링을 출력한다.
```

```python
# 타입 힌트 : 매개변수와 반환값이 어떤 자료형인지 표시하는 기능
# 타입 힌트는 에러가 나도 오류가 나지 않는다.
# def 환률변환(원화:float, 환율:float) -> float: # :float 타입힌트
# 바이브 코딩시 무조건 사용해야 하는 기능임.
def 평균(점수목록:list[int]) -> float: #점수목록 : [10,9,6,4]
    return sum(점수목록) / len(점수목록)

평균([10,9,6,4])
```
