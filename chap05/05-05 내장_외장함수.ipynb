```python
# 내장 함수
# filter(F, Iterable) : F는 함수, Iterable은 반복 가능한 객체. F가 True인 것만 반환
# 요건 : 리스트에서 양수만 추출
datas = [1, -3, 2, 0, -5, 5]
result = [] # 결과를 저장할 빈 리스트 생성
for data in datas:
    if data > 0: # 양수인 경우
        result.append(data) # 결과 리스트에 추가    
print(result) # [1, 2, 5]
```

```python
# filter(조건이 들어있는 함수, 리스트) 사용코드변경
def positive(x): # x:단일값
    return x > 0 # return boolean(True/False) : 비교연산자
# datas = [1, -3, 2, 0, -5, 5]
```

```python
list(filter(positive, datas)) # Iterable filter object, Iterable 놈들은 리스트 변환가능
```

```python
# lambda 명령 : 함수 오브젝트 생성
# 기본 구조 -> lambda 매개변수 ,,,,, : 실행코드

list(filter(lambda x: x > 0, datas)) # lambda로 함수 오브젝트 생성
```

```python
# (lambda a,b,a+d)3,4)_
aaa = lambda a,b: a+b
print(aaa(3,4)) # 7
```

```python
# map(function, iterable) : function 가공, 리스트
# 요건 : 리스트의 값들을 2배 처리 결과
datas = [1, 2, 3, 4] #=> [2,4,6,8]
result = [] # 최종 결과를 저장할 빈 리스트 생성
for data in datas:
    result.append(data*2) # 리스트에 2배 처리한 결과 추가
print(result) # [2, 4, 6, 8]
```

```python
# map 사용
def two_times(x): # x: 단일값 , map 함수가 리스트에서 하나씨 꺼내서 이 함수를 호출    
    return x*2
list(map(two_times, datas)) # map object, 리스트로 변환가능
```

```python
# map 사용
list(map(lambda x: x*2, datas)) # map object, 리스트로 변환가능
```

```python
def plus_one(x):
    return x+1  
```

```python
# 2배
print(list(map(two_times, datas))) # [2, 4, 6, 8]
print(list(map(plus_one, datas))) # [2, 4, 6, 8]
# 1 더하기
```

```python
# zip function
a_data = [1, 2, 3]
b_data = [4, 5, 6]

print(list(zip(a_data, b_data)))
result = zip(a_data, b_data)
for data in result:
    print(data)

```

```python
#pickle 예제
# object = file save
# file W mode open
import pickle   
f = open('student.data', 'wb') # wb : 바이너리 파일 쓰기 모드
data = {1: '홍길동', 2: '고길동', 3: '마이꼴'} # 딕셔너리 오브젝트(바이너리) 생성
pickle.dump(data, f) # object -> file
f.close() # 파일 닫기
```

```python
# file -> object
import pickle
f = open('student.data', 'rb') # rb : 바이너리 파일 읽기 모드   
data = pickle.load(f) # file -> object
f.close() # 파일 닫기
print(data) # 데이터 출력   
print(type(data)) # 데이터 타입 출력
```
