```python
# 클래스 : 비슷한 성격의 데이터(변수)와->self 기능(메소드)을 하나로 묶어
# 오브젝트를 찍어내는 설계도(틀) 입니다.
# 설계도대로 생성된 놈들 -> 객체 : 오브젝트 , 인스턴스
# 반복 사용 의미가 있다 : 함수 비슷하다
# 왜 사용하느가?(함수가 있는데, 왜 클래스가 필요한가?)
# 예를들어 학생 1명에 대한 정보를 저장하려면
# 이름, 점수 1000명 저장한다고 하면,
# 클래스의 목적은 원래 여러개의 변수를 한번에 호출하기 위함
# 요즘은 추가로 클래스로 메소드도 추가된 의미라고 보면 된다

```

```python
# 한명의 학생정보를 저장하는 클래스 선언
# 클래스로 변수와 메소드를 생성한다.
class 학생: # 학생이라는 클래스 선언해라 명령, 이름, 점수 저장하는 공간 선언
    def __init__(self, 이름, 점수): # __init__ : 오브젝트 생성될때 자동 호출되는 메소드
        # self.이름 : 오브젝트 안에 이름 변수 선언, 이름이라는 값을 대입
        self.이름 = 이름 
        # self.점수 : 오브젝트 안에 점수 변수 선언, 점수라는 값을 대입
        self.점수 = 점수 
    def 출력(self):#오브젝트 안에 값을 출력 메소드
        print('self' + '는 오브젝트의 주소값을 의미한다',self)        
        print(f'{self.이름} : {self.점수}')

```

```python
홍길동 = 학생("홍길동", 90) #학생이라는 클래스 ()가 붙으면 오브젝트 생성 
고길동 = 학생("고길동", 80) #학생이라는 클래스 ()가 붙으면 오브젝트 생성 
마이꼴 = 학생("마이꼴", 70) #학생이라는 클래스 ()가 붙으면 오브젝트 생성 

홍길동.출력()
고길동.출력()
마이꼴.출력()
```

```python
# self는 주소값을 의미한다
# object에 접근하려면 self로만 접근이 가능하다
# object에 self가 없다면 접근을 하지 않겠다는 의미
# self값은 변수명에 의해서 self값이 변경된다.
# 변수명.출력() -->변수명.출력(변수명 주소값이 들어간다고 보면 된다.)  가장 중요함.
```

```python
# 클래스를 쓰는 이유
# 1. 데이터(변수)와 기능(메소드)을 묶이
# 2. 재사용 : 설계도 하나로 객체를 무한 생산
# 3. 확장성 : 상속으로 기존 코드를 물려받아 확장
# 4. 현실 모델링 : 현실의 사물을 자연스럽게 표현 -> 자동차 , 스마트폰
# 이세상의 모든 물건은 클래스로 만들수 있다
# 클래스는 변수/메소드가 있다
```

```python
# 계산기 클래스
class Caluculator:
    def __init__(self):
        self.result = 0
    def adder(self, num):
        self.result += num # result에 num 누적
        return self.result
    def sub(self, num):
        self.result -= num # result에 num 누적
        return self.result  
    
```

```python
# 오브젝트 2개 생성
cal1 = Caluculator()
cal2 = Caluculator()

cal1.adder(3) # cal1의 result에 3 누적
cal1.adder(4) # cal1의 result에 3 누적
cal1.sub(2) # cal1의 result에 3 누적
# cal2.adder(7) # cal1의 result에 3 누적
# cal2.adder(7) # cal1의 result에 3 누적
# cal2.sub(4) # cal1의 result에 3 누적
```

```python
# 학생 정보 저장하고 출력 클래스
# 학생 정보 : 학번, 학과, 이름, 성별
# 출력 : 학번, 학과, 이름
# 클래스의 첫자는 대문자로 시작
class StudentInfo: 
    def __init__(self, 학번, 학과, 이름, 성별): # __init__ : 오브젝트 생성될때 자동 호출되는 메소드
        # self.이름 : 오브젝트 안에 이름 변수 선언, 이름이라는 값을 대입
        self.학번 = 학번
        self.학과 = 학과
        self.이름 = 이름
        self.성별 = 성별
    def printStudentInfo(self):#오브젝트 안에 값을 출력 메소드
        print(f'학번 : {self.학번} : 학과 : {self.학과} : 이름 : {self.이름} ')



```

```python
# 클래스 고급 기능
# 1. 클래스 변수 : 클래스의 변수 
# -> 클래스 소유의 변수(클래스 변수는 self로 만들면 안된다)
# 메소드 밖에다가 만든다
# 용도 : 모든 오브젝트가 공유하는 변수
# 오브젝트 변수 : self로 만든 변수
# 예를 설명 : 인터넷 정보 제공 서비스하는 클래스
# 인터넷 회사니까 고객이 돈이된다 
class Service:
    secret = "영구는 외계인이다" # 클래스 변수
```

```python
# 첫번째 고개 : 오브젝트 한개 생성
pey = Service()
kim = Service()

print(pey.secret)
print(kim.secret)
print(Service.secret)
```

```python
Service.secret = "영구는 도깨비이다" # 클래스 변수는 클래스 이름으로 접근해서 변경해야 한다.
print(pey.secret)
print(kim.secret)
```

```python
# 합을 구해주는 서비스를 추가
class Service:
    secret = "영구는 외계인이다" # 클래스 변수
    # 현재는 self가 필요없다. 단, pey.sum(1, 1) 호출될때 self가 필요하다
    # def sum(a, b): # 합을 구하는 서비스를 고객에게 제공하겠다
    #     result = a + b # result는 지역변수
    #     print(f'{a} + {b} = {result} 입니다.')
    # self.result를 하면 오브젝트에 저장된다
    # 클래스 변수는 공유하기 위함이다.
    def __init__(self,name):
        self.name = name
        
    def sum(self, a, b): # 합을 구하는 서비스를 고객에게 제공하겠다
        result = a + b # result는 지역변수
        print(f'{self.name}님의 요청 : {a} + {b} = {result} 입니다.')        
```

```python
# 메소드에서 self가 없으면 클래스의 메소드로 접근하여야 사용이 가능하다.
pey = Service('pey')
kim = Service('kim')
park = Service('park')
print(pey)
print(kim)
print(park)
pey.sum(1, 1) 
kim.sum(1, 1) 
park.sum(1, 1) 
# Service.sum(pey, 1, 1) 
# Service.sum(kim, 1, 1) 
# Service.sum(park, 1, 1) 
```

```python
# self를 받는 메소드에서만 오브젝트의 변수 선언 가능
class Service:
    secret = "영구는 외계인이다"

    def __init__(self, name):
        self.name = name       

    def setName (self, name):
        self.name = name

    def sum(self, a, b):
        result = a + b
        print(f'{self.name}님 {a} + {b} = {result} 입니다.')
        # print(f'님 {a} + {b} = {result} 입니다.')
```
