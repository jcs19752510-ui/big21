```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def speak(self):#상속받는 클래스는 speak 메소드를 반드시 오버라이딩을 해야 한다.
        pass

class Dog(Animal):
    def speak(self):#상속받는 클래스는 speak 메소드를 반드시 오버라이딩을 해야 한다.
        return("멍멍")
    
class Cat(Animal):
    def speak(self):#상속받는 클래스는 speak 메소드를 반드시 오버라이딩을 해야 한다.
        return("야옹")
    
class Cow(Animal):
    def speak(self):#상속받는 클래스는 speak 메소드를 반드시 오버라이딩을 해야 한다.
        return("음메")    
# dos = Dog()

class Duck(Animal):
    def speak(self):#상속받는 클래스는 speak 메소드를 반드시 오버라이딩을 해야 한다.
        return("꽥")
    
duck = Duck()
annimls = [Dog(), Cat(), Cow(),Duck()]

for animal in annimls:
    print(animal.speak())

    
```

```python
# 오류를 처리하는 클래스
class MyError(Exception): # Exception -> 기본오류 처리하는 오류를 처리 : 변수,메소드
    # 오류 발생 원인을 저장/출력
    # 저장 : 변수 선언
    def __init__(self):
        self.message = '허용되지 않는 별명입니다.'
    def __str__(self): # 출력하고 하는 변수를 반환
        return self.message

def say_nick(nick): # nick 받아 처리하는 함수, nick =='바보' 에러발생시키고, 아니면 출력
    if nick == '바보':
        raise MyError() # 예외 오브젝트 생성 던지는 놈
    print(nick)

# say_nick("천사") 

try:       
    say_nick("바보")
except MyError as e:
    print(e)
```
