```python
# 클래스 : 비슷한 성격의 데이터(변수)와 기능(메소드)을 하나로 묶어, 
# 오브젝트를 찍어내는 설계도(틀) 입니다.
# 설계도대로 생성된 놈들 -> 객체 : 오브젝트, 인스턴스 
# 반복 사용 의미가 있다 : 함수 비슷하다

# 왜 사용하는가?
# 예를들어 학생 한명의 정보 저장 : 
# 이름, 점수 => 1000명 저장한다고 하면 
```

```python
# 한명의 학생 정보를 저장하는 클래스 선언
class 학생: # 학생이라는 클래스 선언해라 명령, 이름, 점수 저장하는 공간 선언
    def __init__(self, 이름, 점수): # __init__() : 오브젝트가 생성될 때 자동 호출되는 메소드
        # self.이름 : 오브젝트 안에 이름 변수 선언, 이름이라는 값을 대입
        self.이름=이름 
        # self.점수 : 오브젝트 안에 점수 변수 선언, 점수라는 값을 대입
        self.점수=점수 
    def 출력(self): # 오브젝트 안에 값을 출력 메소드
        print(f'{self.이름} : {self.점수}')
```

```python
홍길동 = 학생('홍길동',90) # 학생이라는 클래스 ()가 붙으면 오브젝트 생성 명령
```

```python
홍길동.출력()
```

```python
# 클래스를 쓰는 이유
# 1. 데이터(변수)와 기능(메소드)을 묶기
# 2. 재사용 : 설계도 하나로 객체를 무한 생산
# 3. 확장성 : 상속으로 기존 코드를 물려받아 확장
# 4. 현실 모델링 : 현실의 사물을 자연스럽게 표현 => 자동차, 스마트폰 
```

```python
# 계산기 클래스 
class Calculator:
    def __init__(self):
        self.result = 0
    def adder(self, num):
        self.result += num # result 에 num 누적
        return self.result
    # 빼기 추가
    def sub(self, num):
        self.result -= num
        return self.result
```

```python
# 오브젝트 2개 생성
cal1 = Calculator()
print(cal1.adder(3))
print(cal1.sub(2))
```

```python
# 지금부터 계산기 클래스 
# 요건 : 더하기 기능을 하고, 결과 반환 계산기
# 1. 클래스 이름 : Calculator, 클래스 이름은 해당 클래스의 용도를 설명할 수 있는 이름
# 2. 변수 : 결과 저장, 반환, 처음에 0 초기화되어야 한다. __init__
# 3. 기능(메소드) : 더하기 메소드, result에 더하기한 결과, 더할 값을 받아야 한다. 
class Calculator:
    def __init__(self): # 결과 저장 변수가 처음에 0으로  초기화
        self.result=0
    def adder(self, number): # result + number -> return result
        # self.result = self.result + number
        self.result += number
        return self.result
```

```python
# 요건 : 더하기 기능을 하고, 결과 반환 계산기
cal1 = Calculator()
print(cal1.adder(3))
print(cal1.adder(4))
```
