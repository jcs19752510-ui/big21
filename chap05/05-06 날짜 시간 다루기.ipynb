```python
# time module , datetime module
# time : 날짜 , 시간 다루는 기능
# 1. time() : 1970. 1. 1. 0:0:0(유닉스타임)를 기준으로 진잔 시간을 초 단위 값을 리턴

import time
print(time.time())

```

```python
import time

# 현재 시간을 가지고 와서 년월일 시분초 분리
# time.localtime() : 현재 시간을 년월일 시분초로 분리해서 튜플로 리턴   
dt = time.localtime() # 분리만 해 준 것이다.
print(dt)
print(dt.tm_year)
print(dt.tm_mon)
print(dt.tm_mday)
print(dt.tm_hour)
print(dt.tm_min)
print(dt.tm_sec)
```

```python
# 날짜 시간 출력
# strftime(포맷,없으면) : 현재 유닉스타임 값을 포맷에 맞게 출력
print(time.strftime('%j'))
print(time.strftime('%u'))
print(time.strftime('%Y'))
print(time.strftime('%Y-%m-%d %H:%M:%S')) # 년-월-일 시:분:초
```

```python
# datetime module : 날짜 시간 다루는 기능
# 클래스 : datetime, date, time, timedelta , (timezone / zoneinfo) 5개 클래스가 있다.
# date :  날짜만 처리하는 클래스
# time :  시간만 처리하는 클래스
# datetime :  날짜 + 시간을 모두 처리하는 클래스
# timedelta : 기간/간격 두 날짜 또는 시간의 차이를 처리하는 클래스(차이를 구할때 사용함)
# timezone / zoneinfo :  시간대 , 한국 + 09:00(KST) + 9시간 의미함.

from datetime import datetime
```

```python
# 특정 날짜와 시간으로 생성
from datetime import date, time, datetime
d = date(2026, 6, 2) # 날짜 오브젝트 생성
print(d)
t = time(15,44,0)
print(t)
dt = datetime(2026, 6, 2) # 시분초 생갸 : 00:00:00
print(dt)
```

```python
# 현재 시각 가져오기 (많이 사용)
from datetime import datetime, date , timezone

# 현재 시각 (날짜 + 시간) , 현재 사용하는 컴퓨터의 시간정보
print(datetime.now(timezone.utc)) # UTC 시간으로 출력
print(date.today()) # 오늘(현재) 날짜
```

```python
# 3. 속성(클래스 안에 있는 변수) 꺼내기
# 속성 : get/set 할 수 있는 변수
# 클래스 안에 있는 변수를 속성이라고 명칭 변경(클래스 public 변수는 그대로 변수라고 부름)
dt = datetime(2026, 6, 2, 15)
print(dt)
print(dt.year, dt.month, dt.day)
print(dt.hour, dt.minute, dt.second)
print(dt.weekday())    # 요일이 숫자로 출력 월:0, 화:1, 수:2, 목:3, 금:4, 토:5, 일:6
print(dt.isoweekday()) # 요일이 숫자로 출력 월:1, 화:2, 수:3, 목:4, 금:5, 토:6, 일:7
```

```python
요일 = ['월', '화', '수', '목', '금', '토', '일']
dt = datetime(2026, 6, 2)
print(f'{dt:%Y-%m-%d}은 {요일[dt.weekday()]}요일') #dt를 필터링 해라.
```

```python
# 문자열 <-> datetime object
# 1. datetime -> 문자열 : strftime() 메서드 사용
# format string을 요즘은 많이 사용한다.
dt = datetime(2026, 6, 2, 15,30,0)
print(dt.strftime('%Y-%m-%d')) # 년-월-일 시:분:초
print(f'{dt:%Y-%m-%d}') # 년-월-일

# 2. 날짜 시간형태로 된 문자열 -> datetime object : strptime()
# 오브젝트로 생성해서 꺼내쓰기 쉽게 하기 위함.
dt_str = '2026/06/02 15:30'
# 시간을 포맷에 맞게 넣어주고 속성으로 꺼내쓰기 위함
dt = datetime.strptime(dt_str, '%Y/%m/%d %H:%M')
print(dt.year)
print(dt.month)
print(dt.day)
#년도 추출
# print(dt_str.split()[0].split('/')[0])
# print(dt_str.split()[0].split('/')[1])
# print(dt_str.split()[0].split('/')[2])
```

```python
# 날짜 계산 : timedelta 
from datetime import datetime, timedelta

now = datetime(2026, 6, 2, 15, 0) # 2026-06-02 15:00:00
# 2026-06-02 15:00:00
# print(now)
# 일주일 뒤를 알고 싶다.
print(now + timedelta(days=7)) # 7일 뒤

# 3시간전 계산
print(now - timedelta(hours=3)) # 3시간 전

# 2주 뒤
print(now + timedelta(weeks=2)) # 2주 뒤

# 복합 : 1일 6시간 30분 뒤
print(now + timedelta(days=1, hours=6, minutes=30)) 
```

```python
# 두 날짜시간 오브젝트의 차이 => timedelta
d1 = datetime(2026, 6, 2)
d2 = datetime(2026, 12, 31)
diff = d2 - d1
print(d1)
print(d2)
print(type(diff)) # <class 'datetime.timedelta'>  
print(diff)
print(diff.days)
```

```python
# 비교와 정렬
d1 = datetime(2026, 6, 2, 9)
d2 = datetime(2026, 6, 2, 18)
print(d1 > d2)
print(d1 < d2)
dates = [datetime(2026, 6, 2), datetime(2026, 6, 1), datetime(2026, 6, 3)]
print(sorted(dates))
```
