```python
# 함수를 쓰레드로 실행(동시 실행)

import threading # 쓰레드 모듈
import time

# 쓰레드는 여러개 아귀먼트를 받을수 있기에 
# 튜플로 매개변수를 받아야 한다. (args=튜플)
# main 쓰레드가 기본으로 1개 존재한다.(스텍 영역에 만들어진다.)
# 쓰레드를 실행하는 main 쓰레드는 
# 다른 쓰레드가 실행되는 동안에도 계속해서 실행된다.
# main 쓰레드가 t1과 sleep lock 넘김 maindmfh
# main 쓰레드가 t2 쓰레드를 실행 sleep lock 넘김 main
# main 다음줄 바로 실행
# main 쓰레드는 종료
# t1 쓰레드 실행 후 종료
# t2 쓰레드 실행 후 종료
# 모드 쓰레드가 종료되야 프로세스가 종료되는 것이다.
def 인사하기(이름:str)->None:
    print(f"{이름}님, 시작")
    time.sleep(2) #wait 개념으로 락을 사람에게 넘긴다는 의미. 2초간 wait. lock 해제
    print(f"{이름}님, 완료!")

# 1. 함수를 이용해서 쓰레드를 생성 (target=실행할함수지정, args=인자값)    
# 1. 철수의 인사하기 쓰레드 생성
# main 쓰레드가 t1과 t2 쓰레드를 생성하는 과정
# t1/t2 쓰레드 종료하기 전에는 main 종료하지 마세요.코드수정
t1 = threading.Thread(target=인사하기, args=("철수",), daemon=False) # 데몬 쓰레드로 만들고 싶으면 True, 일반 쓰레드로 만들고 싶으면 False
# 2. 영희의 인사하기 쓰레드 생성
t2 = threading.Thread(target=인사하기, args=("영희",), daemon=False) # 데몬 쓰레드로 만들고 싶으면 True, 일반 쓰레드로 만들고 싶으면 False

# 쓰레드 시작(start())
t1.start()
t2.start()

# print("모든 작업 완료")

# 3) 두 쓰레드가 끝날때까지 메인 쓰레드를 기다리게 만들어야 한다.
# main을 접근할 수 있는 방법이 없다.
# 그래서 자식인 t1과 t2가 끝날때까지 main이 기다리게 하는 방법이 join()이다.
# t1,t2 쓰레드는 접근이 가능하기에 t1과 t2로 접근하여 통제한다.
# 현재 main이 실행하고 있기에, main이 main을 조정할 수 없다는 의미로 이해하면 됨
# 쓰레드의 wait가 queue 방식이 아니다. sleep이 다른수 있다.
# main은 전체를 마무리하는 역할로 이해하면 된다.
t1.join() # GIL을 가진다는 의미로 보면됨
t2.join() # GIL을 가진다는 의미로 보면됨

print("모든 작업 완료")
```

```python
# 클래스로 쓰레드 생성(상태를 저장을 가질때 사용)
import threading
import time

# Thread 상속받은 다운로드 라는 클래스는 쓰레드가 된다.
# 1. 상속의 의미는 부모가 쓰레드면 나도 쓰레드다
# 2. run() 메서드가 있으면 쓰레드이다
class 다운로드(threading.Thread): 
    def __init__(self, url:str):
        super().__init__() # super == Thread클래스의 __init__() 오브텍트 생성됨. 쓰레드는 부모가 있어야 자식이 만들어 진다.
        self.url  = url    # 쓰레드가 자기 상태를 저장        
        self.상태 = None   # 쓰레드가 자기 상태를 저장
        self.daemon = True  # 데몬 쓰레드로 설정(True / False) (메인이 끝나면 강제로 함께 종료되는 쓰레드)

    # t1.start() -> thread object 안에 있는 run() 호출한다.
    # 쓰레드 처리되야 하는 클래스 반드시 run() 메서드 존재
    # run 메소드 안에 있는 코드가 쓰레드로 작동
    def run(self)->None: 
        print(f"다운로드 시작{self.url} ")
        time.sleep(1)
        self.상태 = f"다운로드 완료{self.url}의 데이터 100KB"


# 다운로드 3개의 쎄륻 생성
ops = [다운로드(f"https://api.aaa.com/{i}") for i in range(3)]

# 쓰레드 3개를 start() 메서드로 쓰레드 시작
for op in ops:
    op.start()

# 쓰레드 3개를 join() 메서드로 쓰레드가 끝날때까지 기다리기
for op in ops:
    op.join()   

# 쓰레드 3개를 쓰레드가 끝난 후 상태 출력
for op in ops:
    print(op.상태)
    
print("다운로드 완료")

```

```python
# 동기화 : 서로 맞춘다.
# 문제상황 : 경쟁상태(race condition) - 여러 쓰레드가 하나의 자원을 동시에 접근할 때 발생하는 문제  
# 쓰레드는 메모리를 공유하므로, 같은 변수를 동시에 건드리면 값이 깨집니다.
# 한개의 은행 계좌 , 동시에 두명이 계좌에서 처리
import threading
import time 

잔액 = 0 # Gloval Var : 공유변수

# 쓰레드로 작동될 함수
def 입금()->None:
    global 잔액 # 공유변수는 global로 선언해야 한다.
    for _ in range(50_000): # 5만번 반복
        현재 = 잔액          # 현재 잔액 가져온다.
        time.sleep(0)
        잔액 = 현재 + 1      # 1원씩


t1 = threading.Thread(target=입금) # 쓰레드 생성
t2 = threading.Thread(target=입금) # 쓰레드 생성

# ts = [threading.Thread(target=입금) for _ in range(2)] # 쓰레드 2개 생성

t1.start() # 쓰레드 시작
t2.start() # 쓰레드 시작

t1.join() # 쓰레드 종료(다 끝난 다음에 종료)
t2.join() # 쓰레드 종료(다 끝난 다음에 종료)

print(f"기대값 : {100000}")  
print(f"실제값 : {잔액}")  

```

```python
# 동기화 : 서로 맞춘다.
# 문제상황 : 경쟁상태(race condition) - 여러 쓰레드가 하나의 자원을 동시에 접근할 때 발생하는 문제  
# 쓰레드는 메모리를 공유하므로, 같은 변수를 동시에 건드리면 값이 깨집니다.
# 한개의 은행 계좌 , 동시에 두명이 계좌에서 처리
import threading
import time 

잔액 = 0 # Gloval Var : 공유변수

# 자물쇠 생성
lock = threading.Lock() # 1. 동기화를 위한 자물쇠 객체 생성(한번에 한 쓰레드만 진입, 재 진입이 안된다.)

# 동기화 : 공유변수 간소화

# 고급지게 배우자
# 속도가 느린게 가장 큰 단점이지만

# 아래 6가지 동기화 도구는 상황에 맞게 사용하면 된다.
# lock    : 한번에 한 쓰레드만 진입, 재 진입이 안된다.
# RLock()   : 재진입 가능
# Semaphore : 개수 제한
# Event    :  신호 보내기 -> 한 쓰레드가 '준비완료' 신호를 주고 다른 쓰레드가 기다리게끔 만드는 것
# Condition :  조건 대기 -> 특정 조건이 될때까지 대기/통지
# Barrier :  집결 지정 -> N개 쓰레드가 모두 도착할때까지 대기시키는게.(예.유튜브)

# 쓰레드로 작동될 함수
def 입금()->None:
    global 잔액 # 공유변수는 global로 선언해야 한다.
    for _ in range(50_000): # 5만번 반복
        with lock:
            현재 = 잔액          # 현재 잔액 가져온다.
            time.sleep(0)
            잔액 = 현재 + 1      # 1원씩

t1 = threading.Thread(target=입금) # 쓰레드 생성
t2 = threading.Thread(target=입금) # 쓰레드 생성

# ts = [threading.Thread(target=입금) for _ in range(2)] # 쓰레드 2개 생성

t1.start() # 쓰레드 시작
t2.start() # 쓰레드 시작

t1.join() # 쓰레드 종료(다 끝난 다음에 종료)
t2.join() # 쓰레드 종료(다 끝난 다음에 종료)

print(f"기대값 : {100000}")  
print(f"실제값 : {잔액}")  

```

```python
# Semaphore : 서버 과부하 , 차단 방지에 필수
# 현재 서버의 용량은 동접 3명까지, 10명이 접속했다.
import threading    
import time

동시접속 = threading.Semaphore(3) # 동시에 접속을 3명까지 허용

def 등록(번호:int)->None:
    with 동시접속:
        print(f"{번호}글카 요청 시작")
        time.sleep(1) # 1년 사용 가장
        print(f"{번호}글카 반납")
threads = [threading.Thread(target=등록, args=(i,)) for i in range(10)] # 10명의 쓰레드 생성
for t in threads:        
    t.start()
for t in threads:
    t.join()

```

```python
# 쓰레드 다루는 고수준(기능이 많다. 간단하게 사용) API : ThreadPoolExecutor(3.2부터 있었음)
# 기본 사용법
# 주식 : 가격, 5종목
from concurrent.futures import ThreadPoolExecutor
import time

def 시세_가져오기(종목:str)->dict:
    time.sleep(1)
    return {"종목":종목, "가격":3000000}

종목들 = ["삼성전자", "하이닉스", "네이버자", "카카오", "현대차"]

with ThreadPoolExecutor(max_workers=5) as pools:
    results = pools.map(시세_가져오기, 종목들) # map() : 입력 순서대로 결과 반환
    for result in results:
        print(result) # result() 메서드로 작업 결과 가져오기
```

```python
# 먼저 끝난 것부터 처리(비동기)
from concurrent.futures import ThreadPoolExecutor, as_completed
import time, random

def 작업(번호:int)->str:
    걸린시간 = random.uniform(0.5, 0.2)
    time.sleep(걸린시간)
    return f"{번호}번 작업 완료, 걸린시간 : {걸린시간:.1f}초"

with ThreadPoolExecutor(max_workers=4) as pools:   
    futures = [pools.submit(작업, i) for i in range(6)] #submit 요청을 보낸것이다.(제출을 완료한게 아니다.)
    for future in as_completed(futures):# 끝나는 순서대로 수신
        try:
           print(f'(완료 : {future.result()}')
        except Exception as e:
            print(f'실패 : {e}')
```

```python
# 쓰레드를 이용해서 데이터를 주고 받는 방법
# 생산자 - 소비자 패턴 : Queue 구조 활용
# Stack : LIFO(Last In First Out)  - 가장 나중에 들어온 데이터가 가장 먼저 나가는 구조 pop()
# Queue : FIFO(First In First Out) - 가장 먼저 들어온 데이터가 가장 먼저 나가는 구조
```

```python
# producer - consumer 패턴 : 생산자와 소비자가 서로 데이터를 주고 받는 패턴
import threading
import queue    
import time

작업큐 = queue.Queue() # 큐생성 , 큐 : 먼저 들어간걸 먼저 나오게 저장소


def 생산자(): # 빵 만드는 사람
    for i in range(5):
        작업큐.put(f'데이터-{i}') # 큐에 작업 추가
        print(f"생산 : 데이터-{i}")
        time.sleep(0.3)    
    작업큐.put(None)

def 소비자(): # 빵 파는 사람
    while True:
        항목 = 작업큐.get() # get()했을때 값이 없을수 있다.
        if 항목 is None: # None이 들어오면 종료
            작업큐.task_done() # 작업 완료 표시
            break
        print(f"소비 : {항목} 처리완료")
        작업큐.task_done() # 작업 완료 표시
```

```python
t1 = threading.Thread(target=생산자)
t2 = threading.Thread(target=소비자)

t1.start()
t2.start()      

t1.join()
t2.join()

print('프로그램 종료')
```
