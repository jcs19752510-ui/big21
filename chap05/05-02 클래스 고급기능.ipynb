```python
# 요건
# 1. 박씨 관리하는 집 : 성이 '박' (클래스 변수로 셋팅)
# 2. fullname : 성 + 이름
# 3. 여행 메소드 : 장소를 입력하면 => "성+이름 , 장소 여행을 가다"
class HousePark:
    lastName = '박' # 클래스 변수

    def __init__(self, name): # 오브젝트 생성될 때 자동으로 호출되는 메소드
        self.getFullName = name # 오브젝트 변수 

    # def setName(self, name): # 오브젝트 변수
    #     self.getFullName = self.lastName + name

    def travel(self, where):
        print(f'{self.getFullName} , {where} 여행을 가다')   

```

```python
aa = HousePark('응용')
print(aa.lastName)
# aa.setName('응용')
print(aa.getFullName)
aa.travel('부산')
```

```python
# 클래스의 상속
# HousePark을 상속받는 HouseKim
# 전제는 기능은 그대로 사용하겠다는 의미
class HouseKim(HousePark):
    lastName = '김' # 클래스 변수

    def travel(self, where):
        print(f'{self.lastName}은 , {where} 여행합니다')         

```

```python
juliet = HouseKim('줄리엣')
print(juliet.lastName)  
juliet.travel('독도')
```
