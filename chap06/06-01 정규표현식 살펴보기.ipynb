```python

data = """
park 800905-1049118
kim 700905-1059119
"""
result = [] #최종결과 저장 빈리스트
# \n -> split -> list
data.split('\n')
for line in data.split('\n'):
    word_result = []
    for word in line.split():
        if len(word) == 14 and word[:6].isdigit() and word[7:].isdigit():
            word_result.append(word[:6] + '-' + '*******')
    result.append(word_result)  
    print(result)
```

```python
import   re

data = """
park 800905-1049118
kim 700905-1059119
"""


pat = re.compile(r"(\d{6})[-](\d{7})")
print(pat.sub(r"\g<1>-*******", data))
```
