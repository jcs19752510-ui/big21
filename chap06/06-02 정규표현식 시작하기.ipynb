```python
# 정규식을 이용한 문자열 검색
# compile :  같은 패턴을 반복 사용할 경우 컴파일 한다
# (1번 사용할 경우 컴파일 할 필요가 없다)

import re #regex 모듈
pat_str = '[a-z]+' #알파벳 소문자 1자이상 패턴
pat = re.compile(pat_str) # 패턴 오브젝트 생성 -> 메소드들 사용할 수 있다.
# re.match() # 1번만 사용할때 사용함.
m1 = pat.match('python') # 여러번 사용할때 사용함. m : mathch 오프젝트 / None
print(m1) # 매치된 문자열 반환
m2 = pat.match('3 python') # 여러번 사용할때 사용함. m : mathch 오프젝트 / None
print(m2) # 매치된 문자열 반환
m3 = pat.search('3 python') # 여러번 사용할때 사용함. m : mathch 오프젝트 / None
print(m3) # 매치된 문자열 반환
```

```python
# 언제 사용
# match : 이 줄이 패턴으로 시작하니?
# search : 이 안에 패턴이 있니?
# findall : 전부 찾아줘
# 매치보다는 통과라는 의미로 받아 들여야 한다
ptr_str = '[a-z]+' #알파벳 소문자 1자이상 패턴
ptr = re.compile(ptr_str) # 패턴 오브젝트 생성 -> 메소드들 사용할 수 있다.
result = ptr.findall('life is too short')
print(result) # 매치된 문자열 반환
```

```python
ptr_str = '[a-z]+' #알파벳 소문자 1자이상 패턴
ptr = re.compile(ptr_str) # 패턴 오브젝트 생성 -> 메소드들 사용할 수 있다.
result = ptr.finditer('life is too short') # 매치된 문자열 반환
print(result) # 매치된 문자열 반환
for res in result:
    print(res) # 매치된 문자열 반환
    print(res.group()) # 매치된 문자열 반환
```
